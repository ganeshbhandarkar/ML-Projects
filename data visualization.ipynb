```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line Plot 

x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)

```

```python
themes = plt.style.available
plt.style.use("seaborn")
print(themes)
plt.plot(x,y1,color="red",label = "Apple",marker="o")
#plt.show()  # breaks and make graph visible
plt.plot(x,y2,color="blue",label = "Orange",linestyle = "dashed",marker="*")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.legend()
plt.show()
```

# Scatter Plot

```python
plt.figure(figsize = (5,10))
plt.scatter(x,y1,color="red",label = "Apple",marker="o")
#plt.show()  # breaks and make graph visible
plt.scatter(x,y2,color="blue",label = "Orange",linestyle = "dashed",marker="*")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot : Prices of Fruits over Time")
plt.legend()
plt.show()
```
