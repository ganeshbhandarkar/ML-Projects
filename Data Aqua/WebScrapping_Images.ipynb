```python
import requests as rs
import bs4
```

```python
url = "https://www.passiton.com/inspirational-quotes?page=3"
```

```python
response = rs.get(url)
```

```python
response.content
```

```python
soup = bs4.BeautifulSoup(response.content)
```

```python
soup
```

```python
article_element = soup.findAll('div',id = "all_quotes")
```

```python
article = article_element[0]
```

```python
article.img.attrs['src']
```

```python
with open('inspir.jpg','wb') as file: # wb write in binary form 
    img_url = article.img.attrs['src']
    response = requests.get(img_url)
    file.write(response.content)
```

```python
for i, article in enumerate(article_element):
    with open('inspir{}.jpg'.format(i),'wb') as file: # wb write in binary form 
        img_url = article.img.attrs['src']
        response = requests.get(img_url)
        
        file.write(response.content)
```
